```
import csv

import numpy as np
import pandas as pd
from scipy.spatial import distance
```

```
# 各ベクトルは(気温,湿度,気圧)
#テストデータ  テストデータは2020/4/3の東京の天気
test_data = np.array([12.8,50,1016.8]).reshape(-1, 1)
```

```
# 各ベクトルは(気温,湿度,気圧)
# 晴れのベクトル　教師データは2020/4/2の東京の天気
sunny_vector = np.array([13.9,45,1003.7]).reshape(-1, 1)
# 雨のベクトル 教師データは2020/4/1の東京の天気
rainy_vector = np.array([11.4,97,1003.9]).reshape(-1, 1)
# 曇りのベクトル 教師データは2020/4/5の東京の天気
cloudy_vector = np.array([10.2,73,1008.4]).reshape(-1, 1)
```

```
#テストデータと晴れのデータの距離
x = distance.euclidean(test_data,sunny_vector)
x = ('{:.2f}'.format(x))
#テストデータと雨のデータの距離
y = distance.euclidean(test_data,rainy_vector)
y = ('{:.2f}'.format(y))
#テストデータと曇りのデータの距離
z = distance.euclidean(test_data,cloudy_vector)
z = ('{:.2f}'.format(z))
print("晴れの教師データとの距離は"+str(x))
print("雨の教師データとの距離は"+str(y))
print("曇りの教師データとの距離は"+str(z))
```

```
if x < y:
    print("予報は晴れです！")
elif y < z:
    print("予報は雨です！")
else:
    print("予報は曇りです！")
```

```
# csvを読み込み、使いやすいように加工
weather_data = pd.read_csv('/content/drive/My Drive/G6 (2年前期)/データと数理I/weather.csv', encoding='shift-jis', header=2)
weather_data.drop(index=[0], inplace=True)
# 余計な列を削除
weather_data.drop(['年月日', '平均気温(℃).1', '平均気温(℃).2', '日照時間(時間).1', '日照時間(時間).2', '日照時間(時間).3', '平均現地気圧(hPa).1', '平均現地気圧(hPa).2', '平均湿度(％).1', '平均湿度(％).2', '平均雲量(10分比).1', '平均雲量(10分比).2', '天気概況(昼：06時〜18時).1', '天気概況(昼：06時〜18時).2'], axis=1, inplace=True)
# この月の天気を晴・雨・曇の３つにざっくり分ける(inplace=Trueで元データを書き換える)
weather_data.replace('晴時々曇', '晴', inplace=True)
weather_data.replace('晴後曇', '晴', inplace=True)
weather_data.replace('快晴', '晴', inplace=True)
weather_data.replace('晴一時曇', '晴', inplace=True)
weather_data.replace('晴後一時薄曇', '晴', inplace=True)
weather_data.replace('薄曇後晴', '晴', inplace=True)
weather_data.replace('雨時々曇', '雨', inplace=True)
weather_data.replace('雨一時曇', '雨', inplace=True)
weather_data.replace('雨後曇', '雨', inplace=True)
weather_data.replace('薄曇', '曇', inplace=True)
weather_data.replace('薄曇一時晴', '曇', inplace=True)
weather_data.replace('曇後一時雨', '曇', inplace=True)
weather_data.replace('曇時々雨後一時晴', '曇', inplace=True)
weather_data.replace('曇後晴', '曇', inplace=True)
weather_data.replace('曇時々雨', '曇', inplace=True)
weather_data.sort_values('天気概況(昼：06時〜18時)', inplace=True)
weather_data
```

```
def forcast_weather(temperature, daylight_hours, pressure, humidity, cloud_cover):
    # 各ベクトルは(平均気温(℃) , 日照時間(h), 平均現地気圧(hPa), 平均湿度(％), 平均雲量(10分比))
    # 晴のベクトル　2019/04/01~2019/04/30の晴の日の平均
    sunny_vector = np.array([13.0, 10.0, 1009.6, 52.9, 4.1]).reshape(-1, 1)
    # 曇のベクトル 2019/04/01~2019/04/30の曇の日の平均
    cloudy_vector = np.array([15.7, 4.0, 1012.5, 67.4, 9.2]).reshape(-1, 1)
    # 雨のベクトル 2019/04/01~2019/04/30の雨の日の平均
    rainy_vector = np.array([10.4, 0.13, 1006.8, 89.5, 10]).reshape(-1, 1)

    # 推定したい天気のデータ
    today_weather = np.array([temperature, daylight_hours, pressure, humidity, cloud_cover]).reshape(-1, 1)

    # 各天気との距離を測る
    sunny_distance = distance.euclidean(today_weather, sunny_vector)
    rainy_distance = distance.euclidean(today_weather, rainy_vector)
    cloudy_distance = distance.euclidean(today_weather, cloudy_vector)

    if sunny_distance < cloudy_distance and sunny_distance < rainy_distance:
        # それぞれの天気ベクトルとの距離を割合で表し近ければ近いほど確率が高いと定義しておく
        sunny_probabillity = (1 - sunny_distance / (sunny_distance + rainy_distance + cloudy_distance)) * 100
        sunny_probabillity = round(sunny_probabillity)
        print('この日の天気は', sunny_probabillity, '%の確率で晴れです')
    elif rainy_distance < sunny_distance and rainy_distance < cloudy_distance:
        rainy_probabillity = (1 - rainy_distance / (sunny_distance + rainy_distance + cloudy_distance)) * 100
        rainy_probabillity = round(rainy_probabillity)
        print('この日の天気は', rainy_probabillity, '%の確率で雨です')
    else:
        cloudy_probabillity = (1 - cloudy_distance / (sunny_distance + rainy_distance + cloudy_distance)) * 100
        cloudy_probabillity = round(cloudy_probabillity)
        print('この日の天気は', cloudy_probabillity, '%の確率で曇りです')
```

```
# 2020/04/27のデータを使用
forcast_weather(12.8, 1.4, 1013.4, 78, 10)
```
